```python
import matplotlib.pyplot as plt
```

```python
f_grpc = open('grpc.log', 'r')
f_node = open('node.log', 'r')
f_java = open('java.log', 'r')
```

```python
grpc = []
for line in f_grpc:
    grpc.append(int(line.split(':')[-1]))

node = []
for line in f_node:
    node.append(int(line.split(':')[-1]))

java = []
for line in f_java:
    java.append(int(line.split(':')[-1]))
```

```python
plt.plot(node)
```
